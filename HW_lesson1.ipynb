### 1. Импорт библиотек

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
%matplotlib inline

from pylab import rcParams
rcParams["figure.figsize"] = 9, 14
```

### 2. Создание множества значений **precision** и **recall**

```python
precision = np.linspace(0.1, 1, 10)
recall = np.linspace(0.1, 1, 10)

precision, recall = np.meshgrid(precision, recall)
```

### 3. Определение арифметических и гармонических средних

```python
mean = (precision + recall) / 2
f = 2 * (precision * recall) / (precision + recall)
```

### 4. Создание иллюстраций

```python
fig = plt.figure(figsize=plt.figaspect(0.5))

# Поверхность средних значений
ax = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax.plot_surface(precision, recall, mean, rstride=1, cstride=1, cmap=cm.jet,
                       linewidth=0, antialiased=False)
ax.set_zlim(0, 1)

ax.set_xlabel("Precision")
ax.set_ylabel("Recall")
ax.set_zlabel("Среднее значение")
ax.set_title("Поверхность средних значений")

# Поверхность гармонических значений
ax = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(precision, recall, f, rstride=1, cstride=1, cmap=cm.jet,
                       linewidth=0, antialiased=False)
ax.set_zlim(0, 1)

ax.set_xlabel("Precision")
ax.set_ylabel("Recall")
ax.set_zlabel("F-мера")
ax.set_title("Поверхность гармонических средних значений")
```

### Вывод
Как видно из данных иллюстраций, среднее значение не показывает связи между **precision** и **recall**. Так при увеличение только одной характеристики ведёт к значительному росту среднего. Использование F-меры помогает оценивать вклад каждой метрики (**precision** и **recall**) в общую оценку. Например увеличение **precision**, при фиксированном значении **recall**, не даст значительного прироста к F-мере.
